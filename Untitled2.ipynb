```python
# Write a Python function factorial(num) which returns the factorial of a given number.
def factorial(num):
    fact = 1
    for i in range(1, num + 1):
        fact *= i
    return fact

# Invoke the function
number = int(input("Enter a number: "))
result = factorial(number)

# Display the result
print("The factorial of", number, "is:", result)
```

```python
'''Write a Python Function is_palindrome(num) that accepts an integer num as argument
 and returns True if the num is palindrome else returns false.
  Invoke the function and based on return value print the output.'''
  def is_palindrome(num):
    original = num
    reverse = 0

    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num //= 10

    return original == reverse

# Invoke the function
number = int(input("Enter a number: "))

if is_palindrome(number):
    print("The number is a palindrome")
else:
    print("The number is not a palindrome")
```

```python
'''Write a Python function check_amicable_numbers(num1, num2) that accepts two numbers num1 and num2 as arguments
and returns True if the given pair of numbers are amicable numbers else return false.
Invoke the function and based on return value print the numbers are amicable numbers or not.
num1 and num2 are said to be amicable numbers if sum of all the proper devisors (except num1 itself) of num1 is equal to num2
 and sum of all the proper devisors of num2 (except num1 itself) is equal to num1.'''
def check_amicable_numbers(num1, num2):
    sum1 = 0
    sum2 = 0

    # Sum of proper divisors of num1
    for i in range(1, num1):
        if num1 % i == 0:
            sum1 += i

    # Sum of proper divisors of num2
    for i in range(1, num2):
        if num2 % i == 0:
            sum2 += i

    return sum1 == num2 and sum2 == num1


# Invoke the function
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if check_amicable_numbers(a, b):
    print("The numbers are amicable numbers")
else:
    print("The numbers are not amicable numbers")
```

```python
'''Write a Python function right_shift(num, n) that takes two numbers num and n as  arguments
and returns value of the integer num rotated to the right by n positions.
Assume both the numbers are unsigned. Invoke the function and print the return value.

Hint: use >> binary operator to shift the bits.'''
def right_shift(num, n):
    bits = 32  # assuming 32-bit unsigned integer
    n = n % bits
    return (num >> n) | (num << (bits - n)) & (2**bits - 1)


# Invoke the function
num = int(input("Enter number: "))
n = int(input("Enter number of positions to rotate: "))

result = right_shift(num, n)
print("Result after right rotation:", result)
```

```python
'''Write a Python function check_strong_number(num) that accepts a positive integer as argument and
returns True if the number is strong number
 else false. Invoke the function and based on return value print the number is strong number or not.

A number is said to be strong number, if the sum of factorial of each digit of the number is equal to the given number.'''

def check_strong_number(num):
    temp = num
    total = 0

    while temp > 0:
        digit = temp % 10
        fact = 1
        for i in range(1, digit + 1):
            fact *= i
        total += fact
        temp //= 10

    return total == num


# Invoke the function
number = int(input("Enter a number: "))

if check_strong_number(number):
    print("The number is a strong number")
else:
    print("The number is not a strong number")
```
